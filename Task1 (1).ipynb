```python
import pandas as pd
import networkx as nx

neighborhoods_df = pd.read_csv('Neighborhoods.csv')
facilities_df = pd.read_csv('Facilities.csv')
existing_roads_df = pd.read_csv('Existing_Roads.csv')
potential_roads_df = pd.read_csv('Potential_Roads.csv')
traffic_flow_df = pd.read_csv('Traffic_Flow.csv')
metro_lines_df = pd.read_csv('Metro_Lines.csv')
bus_routes_df = pd.read_csv('Bus_Routes.csv')
transport_demand_df = pd.read_csv('Transport_Demand.csv')

print("Successfully read all 8 files!\n")
print("-" * 40)

G = nx.Graph()

for index, row in neighborhoods_df.iterrows():
    G.add_node(str(row['ID']), name=row['Name'], type='Neighborhood')

for index, row in facilities_df.iterrows():
    G.add_node(str(row['ID']), name=row['Name'], type='Facility')

for index, row in existing_roads_df.iterrows():
    G.add_edge(str(row['FromID']), str(row['ToID']), weight=row['Distance(km)'])

print("Successfully built the existing road network!")
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
import pandas as pd

neighborhoods_df = pd.read_csv('Neighborhoods.csv')
facilities_df = pd.read_csv('Facilities.csv')
existing_roads_df = pd.read_csv('Existing_Roads.csv')
potential_roads_df = pd.read_csv('Potential_Roads.csv')
traffic_flow_df = pd.read_csv('Traffic_Flow.csv')
metro_lines_df = pd.read_csv('Metro_Lines.csv')
bus_routes_df = pd.read_csv('Bus_Routes.csv')
transport_demand_df = pd.read_csv('Transport_Demand.csv')

neighborhoods_df['ID'] = neighborhoods_df['ID'].astype(str)
facilities_df['ID'] = facilities_df['ID'].astype(str)
existing_roads_df['FromID'] = existing_roads_df['FromID'].astype(str)
existing_roads_df['ToID'] = existing_roads_df['ToID'].astype(str)
potential_roads_df['FromID'] = potential_roads_df['FromID'].astype(str)
potential_roads_df['ToID'] = potential_roads_df['ToID'].astype(str)
transport_demand_df['FromID'] = transport_demand_df['FromID'].astype(str)
transport_demand_df['ToID'] = transport_demand_df['ToID'].astype(str)

traffic_flow_df[['FromID', 'ToID']] = traffic_flow_df['RoadID'].str.split('-', expand=True)
traffic_flow_df['FromID'] = traffic_flow_df['FromID'].astype(str)
traffic_flow_df['ToID'] = traffic_flow_df['ToID'].astype(str)
traffic_flow_df = traffic_flow_df.drop('RoadID', axis=1)

metro_lines_df['Stations'] = metro_lines_df['Stations'].str.split(',')
bus_routes_df['Stops'] = bus_routes_df['Stops'].str.split(',')

print("Data cleaning and preprocessing completed successfully!")
```
